## Imports / functions

```python
import anndata
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
def plot_distribution(x, name, column):
    plt.figure(figsize=(8, 4))
    plt.xticks(rotation=45, fontsize=10)
    sns.histplot(x.obs[column], kde=True)
    plt.title(f'Distribution of {column} in {name}')
    plt.gcf().subplots_adjust(bottom=0.4)
    plt.xlabel(column)
    plt.ylabel('Frequency')
    plt.tight_layout()
    plt.show()
```

## Datasets to process

```python
datasets = {
  'KPMP 1': 'dataset_1.h5ad',
  'KPMP 2': 'dataset_2.h5ad',
#   'HuBMAP Left Kidney': 'hubmap-LK-processed.h5ad',
#   'HuBMAP Right Kidney': 'hubmap-RK-processed.h5ad'
}
```

## Processing

Render distribution plots for suitable variables in each dataset

```python
for (dataset, h5ad) in datasets.items():
    x = anndata.read_h5ad(h5ad)
    for column in x.obs.columns:
        num_values = x.obs[column].nunique()
        # Only render distributions with a small number of categories
        if num_values > 1 and num_values < 50:
            plot_distribution(x, dataset, column)
```
